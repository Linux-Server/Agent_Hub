```python
# read the data
with open('input.txt',  'r', encoding='utf-8') as file : 
    text = file.read()
```

```python
len(text)
```

```python
print(text[:1000])
```

```python

chars = sorted(list(set(text)))
print(len(chars))

print(chars)
```

```python
print(''.join(chars))
```

```python
in_char = {i:ch for i,ch in enumerate(chars)}
char_in = {ch:i for i,ch in enumerate(chars)}

```

```python
encode = lambda x : [char_in[i] for i in x]
decode = lambda x : ''.join([in_char[i] for i in x])

encoded_vector = encode(text)

encoded_vector
```

```python
import torch

data = torch.tensor(encoded_vector, dtype=torch.long)
print(data.shape)

print(data[:1000])
```

```python
# let's build train and validation set
n = int(0.9 * len(data))
train_data = data[:n]
val_data = data[n:]
```
